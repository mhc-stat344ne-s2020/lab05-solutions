```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import math
```

```python
benedict_path = "http://www.evanlray.com/stat344ne_s2020/homework/hw1/benedict.jpg"
benedict = np.array(imageio.imread(benedict_path))[50:1250, 75:1275, :]
benedict_grayscale = np.mean(benedict, axis = 2).astype(int)
print(benedict_grayscale.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (12, 12))
ax.imshow(benedict_grayscale, cmap = "gray")
```

```python
benedict_padded = np.pad(benedict_grayscale, ((0, 0), (0, 0)))
W = np.array([[1, 0, -1],
              [2, 0, -2],
              [1, 0, -1]])
print(W.shape)
```

```python
# specify the dimensions of the output for the filter.  We will not use padding or a stride.
output = np.zeros((benedict_grayscale.shape[0] - 3 + 1, benedict_grayscale.shape[1] - 3 + 1))

# write a pair of nested loops that do the filtering
for i in range(benedict_grayscale.shape[0] - 3 + 1):
  for j in range(benedict_grayscale.shape[1] - 3 + 1):
    output[i, j] = np.sum(W * benedict_padded[i:(i+3), j:(j+3)])
```

```python
# plot your filtered image with the detected edges.
plt.imshow(output, cmap = "gray")
```
